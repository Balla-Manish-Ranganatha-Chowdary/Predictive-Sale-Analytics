## 3. Train–Test Split

```python
daily_sales = daily_sales.sort_values('date')

train = daily_sales[daily_sales['date'] < '2017-01-01']
test  = daily_sales[daily_sales['date'] >= '2017-01-01']

print(train.shape, test.shape)
```

## 4. ARIMA Baseline Model

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
arima_model = ARIMA(train['sales'], order=(5,1,0))
arima_result = arima_model.fit()

# Forecast
arima_forecast = arima_result.forecast(steps=len(test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Actual')
plt.plot(test['date'], arima_forecast, label='ARIMA Forecast')
plt.legend()
plt.show()
```

## 5. Prophet Model

```python
from prophet import Prophet

prophet_train = train.rename(columns={'date': 'ds', 'sales': 'y'})

prophet_model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True
)

prophet_model.fit(prophet_train)

future = prophet_model.make_future_dataframe(periods=len(test))

prophet_forecast = prophet_model.predict(future)
```

```python
prophet_test_forecast = prophet_forecast.tail(len(test))['yhat'].values
```

```python
prophet_model.plot(prophet_forecast)
```

## 6. Machine Learning Model

```python
df_ml = daily_sales.copy()

df_ml['day'] = df_ml['date'].dt.day
df_ml['month'] = df_ml['date'].dt.month
df_ml['year'] = df_ml['date'].dt.year

# Lag features
df_ml['lag_1'] = df_ml['sales'].shift(1)
df_ml['lag_7'] = df_ml['sales'].shift(7)

# Rolling mean
df_ml['rolling_7'] = df_ml['sales'].rolling(7).mean()

df_ml = df_ml.dropna()
```

```python
train_ml = df_ml[df_ml['date'] < '2017-01-01']
test_ml  = df_ml[df_ml['date'] >= '2017-01-01']

X_train = train_ml.drop(columns=['date', 'sales'])
y_train = train_ml['sales']

X_test = test_ml.drop(columns=['date', 'sales'])
y_test = test_ml['sales']
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)
```

## 9. Model Evaluation

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
#ARIMA
arima_rmse = rmse(test['sales'].values, arima_forecast)
arima_mape = mape(test['sales'].values, arima_forecast)
```

```python
#Prophet
prophet_rmse = rmse(test['sales'].values, prophet_test_forecast)
prophet_mape = mape(test['sales'].values, prophet_test_forecast)
```

```python
#ML Model(Random Forest)
ml_rmse = rmse(y_test.values, rf_predictions)
ml_mape = mape(y_test.values, rf_predictions)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["ARIMA", "Prophet", "ML (Random Forest)"],
    "RMSE": [arima_rmse, prophet_rmse, ml_rmse],
    "MAPE (%)": [arima_mape, prophet_mape, ml_mape]
})

results
```

### Model Selection Rationale

The ML-based model achieved the lowest error metrics, indicating its ability to capture nonlinear temporal patterns. Prophet performed well for seasonality but was outperformed by the ML model in overall accuracy. ARIMA served as a baseline.
